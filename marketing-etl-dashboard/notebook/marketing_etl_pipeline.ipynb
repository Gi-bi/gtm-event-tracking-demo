```python
import pandas as pd

df = pd.read_csv('raw_marketing_data.csv')
df.head()
```

```python
df.columns.tolist()
df.info()
df.isna().sum()
```

```python
df.columns = [c.strip().lower().replace(' ', '_') for c in df.columns]
df.head()
```

```python
df.columns = [c.strip().lower() for c in df.columns]
df.columns
```

```python
df['acquisition_cost'] = (
    df['acquisition_cost']
    .astype(str)
    .str.replace(r'[^0-9.]', '', regex=True)
)

df['acquisition_cost'] = pd.to_numeric(df['acquisition_cost'], errors='coerce')
```

```python
df['channel_clean'] = (
    df['channel_used']
    .str.lower()
    .str.replace(r'fb|facebook', 'facebook', regex=True)
    .str.replace(r'ig|instagram', 'instagram', regex=True)
    .str.replace(r'google|adwords|ppc', 'google', regex=True)
    .str.replace(r'email|newsletter', 'email', regex=True)
)
```

```python
df['ctr'] = df['clicks'] / df['impressions']
df['cpc'] = df['acquisition_cost'] / df['clicks']
df['cost_per_conversion'] = df['acquisition_cost'] / (df['conversion_rate'] * df['clicks'])
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.to_csv('clean_marketing_data.csv', index=False)
```
